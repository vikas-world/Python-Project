# If the user has taken both the doses then it should show,
“Congratulations! You are fully vaccinated”.
2. If the user has not taken any dose then it should show, “Please take your
first dose at earliest”.
3. If the user has taken 1st dose but not taken the 2nd dose and if the
difference between 1st dose date and today’s date is greater than 90 days
then it should show, “You have taken your first dose on <date of first
dose>, please take your 2nd dose at earliest.
4. If the user has taken 1st dose but not taken the 2nd dose and if the
difference between 1st dose and today’s date is not greater than 90 days
then it should show, “You have taken your first dose on <date of first
dose>, please take your 2nd dose on or after <next dose date>.”
Note: The next dose will be 90 days after the first dose date

```python
import pandas as pd
from datetime import timedelta

expected_username = 'vikas_world'
password = '1234'
max_attempts = 3

# Authentication loop for username
for attempt in range(max_attempts):
    user_name = input("Enter your username: ")
    if user_name == expected_username:
        break
    else:
        print("Invalid username. Try again.")

# If the username is incorrect after max_attempts, exit
if user_name != expected_username:
    print("Maximum attempts reached. Exiting...")
else:
    # Username is correct, now ask for the password
    for attempt in range(max_attempts):
        entered_password = input("Enter your password: ")
        if entered_password == password:
            print("Login Successful")
            break
        else:
            print("Invalid password. Try again.")
    else:
        # If the loop completes without a break, it means the password attempts exceeded.
        print("Maximum attempts reached. Exiting...")
        
    if entered_password == password:
        def check_vaccination_status(user_id):
            dataset = pd.read_excel("C:\\Users\\vikas\\OneDrive\\Desktop\\Iota Notes\\vaccine_data.xlsx")
            user = dataset.loc[dataset['userid'] == user_id]
        

            if not user.empty:
                name = user['name'].values[0]
                first_dose_status = user['1st dose status'].values[0]
                first_dose_date = user['date of 1st dose'].values[0]
                second_dose_status = user['2nd dose status'].values[0]

                if first_dose_status == 'yes' and second_dose_status == 'no':
                    first_dose_date = pd.to_datetime(first_dose_date)
                    next_dose_date = first_dose_date + timedelta(days=90)
                    first_dose_date_str = first_dose_date.strftime("%d/%m/%Y")
                    next_dose_date_str = next_dose_date.strftime("%d/%m/%Y")

                    return f"You have taken your first dose on {first_dose_date_str}, please take your 2nd dose on {next_dose_date_str}."

                if first_dose_status == 'yes' and second_dose_status == 'yes':
                    return f"Hi, {name}! Congratulations, you are fully vaccinated."

                if first_dose_status == 'yes':
                    return f"Hi, {name}! You are partially vaccinated."

                return f"Hi, {name}! Please take your first dose at the earliest."

            return "User ID not found."

        # Keep asking for user ID until the user logs out (enters 'q')
        while True:
            user_id = input("Enter user ID: ")
            if user_id.lower() == "q":
                print("Logout")
                break
            else:
                result = check_vaccination_status(user_id)
                print(result)
```
