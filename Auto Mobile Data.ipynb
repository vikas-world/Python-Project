# 1. what is the average price and average mileage for convertible cars

```python
convertable_cars = df[df['body-style'] == 'convertible']
```

```python
average_price = convertable_cars['price'].mean()
average_mileage = convertable_cars['average-mileage'].mean()
```

```python
df1 = pd.DataFrame({
    'Average Price': [average_price],
    'Average Mileage': [average_mileage]
})
```

```python
df1
```

# 2. Sort cars by company and price in decending order

```python
df2 = df.sort_values(['company', 'price'] ,ascending= False,ignore_index=True)[['company', 'price']]
```

```python
df2 = pd.DataFrame(df2)
```

```python
df2
```

# 3. Save data for 'Toyota' separately

```python
toyota_cars = df[df['company'] == 'Toyota']
df3 = toyota_cars.to_csv('toyota_cars.csv', index=False)
```

# 4. which company is producing most expensive car?

```python
df4 = df.groupby('company')['price'].max().nlargest(1)
```

```python
df4 = pd.DataFrame(df4)
```

```python
df4
```

# 5. Name the cars on basis of maximum length, highest wheelbase, maximum horsepower

```python
df5 = df.groupby('company')['length','wheel-base','horsepower'].max()
```

```python
df5 = pd.DataFrame(df5)
```

```python
df5
```

# 6. company wise average price of cars and order them in decending order

```python
df6 = df.groupby('company')['price'].mean().sort_values(ascending=False)
```

```python
df6 = pd.DataFrame(df6)
```

```python
df6
```

# 7. create a column named: "decision", If wheel-base is less than 100 and length is less than 175 then "don't buy" else "buy"

```python
df7 = df['decision'] = np.where((df['wheel-base'] < 100) & (df['length'] < 175) ,'don\'t buy','buy')
```

```python
df7 = pd.DataFrame(df)
```

```python
df7
```

# 8. Select only cars whose decision is "buy"; find average price based on company of car

```python
buy_cars = df.groupby('decision').get_group('buy')
```

```python
df8 = average_price = buy_cars.groupby('company')['price'].mean()
```

```python
df8 = pd.DataFrame(df8)
```
