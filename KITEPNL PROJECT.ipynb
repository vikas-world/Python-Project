```python
DF = df[df['Product']== 'MIS'] # we get only MIS product bcoz this is only requried
```

```python
DF
```

```python
DF1 = DF[DF['Status']== 'COMPLETE'] # we get only complete status bcoz this is only requried
```

```python
DF1
```

```python
DF1['Qty.'] = DF1['Qty.'].str.replace('/1000','')
DF1['Qty.'] = DF1['Qty.'].str.replace('/250','')
DF1['Qty.'] = DF1['Qty.'].str.replace('/2000','')
DF1['Qty.']= DF1['Qty.'].astype('int64')


```

```python
DF1
```

```python
DF1['Turnover'] = DF1['Qty.'] * DF1['Avg. price'] # we get turnover to this methode
```

```python
DF1
```

```python
DF1['Brokerage'] = np.where((DF1['Turnover']*0.0003<20)&(DF1['Turnover']*0.0003>20), DF1['Turnover']*0.0003,20) # find the brokerage depends on condition if brokerage percent 0.0003>20 so put this percent value and if the brokerage<20 so we put 20
```

```python
DF1
```

```python
DF1['STT'] = np.where(DF1['Type'] == 'SELL',DF1['Turnover'] * 0.00025, 0) # apply the rule on sell side
```

```python
DF1
```

```python
DF1['NSE'] = (DF1['Turnover'] * 0.00325/100)
```

```python
DF1
```

```python
DF1['SEBI'] = DF1['Turnover']*0.000001+DF1['Turnover']*0.000001
```

```python
DF1
```

```python
DF1['GST'] = DF1['Brokerage']*0.18 +DF1['NSE']*0.18 +DF1['SEBI']*0.18
```

```python
DF1
```

```python
DF1['Stamp charge'] = DF1.apply(lambda row: row['Turnover'] * 0.00003 if row['Type'] == 'BUY' else 0, axis=1)
```

```python
DF1
```

```python
DF1['Total Charges'] = DF1['Brokerage'] + DF1['GST'] + DF1['SEBI'] + DF1['Stamp charge'] +DF1['STT'] + DF1['NSE']
```

```python
DF1
```

```python
DF_2 = DF1.drop(['Time'],axis=1)
```

```python
DF2 = pd.DataFrame(DF_2)
```

```python
DF2
```

# 3) Overall Summary of each StocksInstrument Gross PnL Total charges Net PnL % Charges on Gross PnL



```python
DF4['Gross PnL'] = DF4.groupby('Instrument')['Turnover'].diff()
```

```python
DF4
```

```python
DF4['Total Charge'] = DF4.groupby('Instrument')['Total Charges'].transform('sum')
```

```python
DF4
```

```python
DF5 = DF4[DF4['Type'] != 'BUY']
```

```python
DF6 = DF5.drop(['Qty.','weighted Avg.price','Turnover','Brokerage','STT','NSE','SEBI','GST','Stamp charge','Type','Total Charges'],axis=1)
```

```python
DF6
```

```python
DF6['Net PnL'] = DF6['Gross PnL'] - DF6['Total Charge']
```

```python
DF6
```

```python
DF6['% Charge'] = (DF6['Total Charge'] / DF6['Gross PnL']) * 100
```

```python
DF7 = pd.DataFrame(DF6)
```

```python
DF7
```

```python
x = DF7.sum(numeric_only=True)
x['Instrument'] = 'Total'
DF9 = DF7.append(x, ignore_index=True)
```

```python
DF10 = pd.DataFrame(DF9)
```
